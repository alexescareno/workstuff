```python
###CONNECTING to data using python with the database in your python repository, copy and pasting the path will also work
con = sqlite3.connect("sakila.db")
```

```python
####using a function I learned to turn comments into sql queries
def sql_to_df(sql_query):
    df = pd.read_sql(sql_query, con)
    return df
```

```python
df
```

```python
###SELECTING data and turning it into a data frame using python
query = ''' SELECT first_name,last_name 
            FROM customer; '''
sql_to_df(query).head()
```

```python
###SELECTING continued
query = ''' SELECT *
            FROM customer; '''
sql_to_df(query).head()
```

```python
##using DISTINCT, to not count twice things such as duplicates.
query = ''' SELECT DISTINCT(country_id)
            FROM city'''
sql_to_df(query)
```

```python
##USING FROM, WHERE, AND
query = '''SELECT *
            FROM film
            WHERE release_year = 2006
            AND rating = 'R' '''
sql_to_df(query).head()
```

```python

#### using COUNT
query = '''SELECT COUNT(customer_id)
            FROM CUSTOMER'''
sql_to_df(query).head()
```

```python
###USING LIKE WITH ' % ' or ' _ ' to search for strings in data with the M character in the first_name column

query = '''SELECT * 
           FROM customer
           WHERE first_name LIKE 'M%' '''
```

```python
sql_to_df(query).head()
```

```python
##the same here, again just searching names using LIKE but with ' _ ' 
query = '''SELECT * FROM
            customer WHERE
            last_name LIKE '_ing' '''

sql_to_df(query).head()
```

```python
## Using GLOB to grab a list of names in the first name that start with either A or B
query = ''' SELECT * FROM
            customer WHERE
            first_name GLOB '[AB]*' ; '''

sql_to_df(query).head()
```

```python
#Using ORDER BY to sort a list of last names in descending order
query = '''SELECT *
            FROM customer
            ORDER BY last_name  DESC; '''

sql_to_df(query).head()
```

```python
#Using GROUP BY to do a count using the store ids and counting the customer ids

query = '''SELECT store_id, COUNT(customer_id)
            FROM customer
            GROUP BY store_id;'''

sql_to_df(query).head()
```
